```python
# merge the orders, order_ds, and order_pay together 
orders_combo = (orders.merge(order_ds, on='order_id', how='left')).merge(order_pay, on='order_id', how='left')
orders_combo['price'] = round(orders_combo['price'], 2)
orders_combo['freight_value'] = round(orders_combo['freight_value'], 2)
orders_combo['payment_value'] = round(orders_combo['payment_value'], 2)
orders_combo['price_diff'] = round((orders_combo['price'] + orders_combo['freight_value'] - orders_combo['payment_value']), 2)
orders_combo[orders_combo['price_diff']!= 0] 
```

```python
# find the total cost of each order that was paid for with a voucher
order_pay[order_pay['payment_type']=='voucher'].groupby('order_id')[['payment_value']].sum()
```

After exploring the data, it is clear that the total order_pay combines the item price and freight value into a single payment_value field. As such, it is best to use the orders table and the orders_ds table to bring in customer ids and only bring in voucher values from the order_pay table to calculate customer life-time value at the end. 

```python
orders = orders.merge(order_ds, on='order_id', how='left')
orders
```

```python
# Merge the orders and sellers data sets together on seller ID
orders_sellers = orders.merge(sellers, on='seller_id', how='left')
orders_sellers
```

```python
# Merge the marketing and closed order sets on mql_id
marketing_closed = marketing.merge(closed, on='mql_id', how='left')
marketing_closed
```

```python
# Merge the orders_sellers and marketing_closed data sets on seller_id
brazil = orders_sellers.merge(marketing_closed, on='seller_id', how='left')
brazil
```

```python
# Merge Brazil with cust table on customer_id
brazil = brazil.merge(cust, on='customer_id', how='left')
# Merge brazil with products on order_id
brazil = brazil.merge(products, on='product_id', how='left')
# Merge brazil with product_translation on product_id
brazil = brazil.merge(product_translation, on='product_category_name', how='left')
brazil
```
