```python
# pair down the columns to what is needed
final = brazil[['order_id', 'order_item_id', 'product_id', 'seller_id', 'price', 
                'order_purchase_timestamp',  'customer_unique_id', 'product_category_name_english', 'customer_id']]

# convert the order purchase time stamp to a datetime 
final['order_purchase_timestamp'] = pd.to_datetime(final['order_purchase_timestamp'])

# ignore all data before 2017
final = final[final['order_purchase_timestamp'] >= '2017-01-01']
# sort the df by order_purchase_timestamp
final = final.sort_values(by='order_purchase_timestamp')
# create an order month 
final['order_month'] = final['order_purchase_timestamp'].dt.strftime('%Y-%m')
final
```

```python
# crate a DF to contain order volume
order_volume = final.groupby(['order_month'])[['order_id']].count().reset_index()
order_volume['previous_months_volume'] = order_volume['order_id'].shift(periods=1)

# create a DF to contain monthly revenue
revenue = final.groupby(['order_month'])[['price']].sum().reset_index()
revenue['previous_months_price'] = revenue['price'].shift(periods=1)

# merge the revenue and order volume tables together

history = order_volume.merge(revenue, on='order_month', how='inner')
history.columns=(['order_month','orders_volume', 'previous_months_volume', 'revenue', 'previous_months_revenue'])
history['volume_diff'] = history['orders_volume'] - history['previous_months_volume']
history['revenue_diff'] = history['revenue'] - history['previous_months_revenue']
history
```

```python
cust_values = final[['customer_unique_id', 'order_month']].groupby('order_month')[['customer_unique_id']].nunique().reset_index()
cust_values.rename(columns={'customer_unique_id':'num_customers'}, inplace=True)
cust_values
```

```python
#create a list of unique customer_unique_id for each month in term
jans = set(list(final[final['order_month']=='2017-01']['customer_unique_id']))
febs = set(list(final[final['order_month']=='2017-02']['customer_unique_id']))
mars = set(list(final[final['order_month']=='2017-03']['customer_unique_id']))
aprs = set(list(final[final['order_month']=='2017-04']['customer_unique_id']))
mays = set(list(final[final['order_month']=='2017-05']['customer_unique_id']))
juns = set(list(final[final['order_month']=='2017-06']['customer_unique_id']))
julys= set(list(final[final['order_month']=='2017-07']['customer_unique_id']))
augs= set(list(final[final['order_month']=='2017-08']['customer_unique_id']))
seps= set(list(final[final['order_month']=='2017-09']['customer_unique_id']))
octs= set(list(final[final['order_month']=='2017-10']['customer_unique_id']))
novs= set(list(final[final['order_month']=='2017-11']['customer_unique_id']))
decs= set(list(final[final['order_month']=='2017-12']['customer_unique_id']))
jans18 = set(list(final[final['order_month']=='2017-01']['customer_unique_id']))
febs18  = set(list(final[final['order_month']=='2017-02']['customer_unique_id']))
mars18  = set(list(final[final['order_month']=='2017-03']['customer_unique_id']))
aprs18  = set(list(final[final['order_month']=='2017-04']['customer_unique_id']))
mays18  = set(list(final[final['order_month']=='2017-05']['customer_unique_id']))
juns18  = set(list(final[final['order_month']=='2017-06']['customer_unique_id']))
julys18 = set(list(final[final['order_month']=='2017-07']['customer_unique_id']))
augs18 = set(list(final[final['order_month']=='2017-08']['customer_unique_id']))


# create a function that compares two lists together and pull keeps the shared customer_unique_id
def intersection(lst1, lst2): 
    lst3 = [value for value in lst1 if value in lst2] 
    return lst3 
        
cust_values['retained'] = np.where(cust_values['order_month']=='2017-01', 0, '')
cust_values['retained'] = np.where(cust_values['order_month']=='2017-02', len(intersection(jans,febs)), cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2017-03', len(intersection(febs, mars)), cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2017-04', len(intersection(mars,aprs)), cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2017-05', len(intersection(aprs,mays)), cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2017-06', len(intersection(mays, juns)), cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2017-07', len(intersection(juns,julys)),cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2017-08', len(intersection(julys,augs)),cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2017-09', len(intersection(augs,seps)),cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2017-10', len(intersection(seps,octs)), cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2017-11', len(intersection(octs, novs)), cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2017-12', len(intersection(novs, decs)),cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2018-01', len(intersection(decs, jans18)),cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2018-02', len(intersection(jans18,febs18)), cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2018-03', len(intersection(febs18, mars18)), cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2018-04', len(intersection(mars18,aprs18)), cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2018-05', len(intersection(aprs18,mays18)), cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2018-06', len(intersection(mays18, juns18)), cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2018-07', len(intersection(juns18,julys18)),cust_values['retained'])
cust_values['retained'] = np.where(cust_values['order_month']=='2018-08', len(intersection(julys18,augs18)),cust_values['retained'])
```

```python
history = history.merge(cust_values, on='order_month', how='left')
history = history[history['order_month']!='2018-09']
history['previous_months_cust'] = history['num_customers'].shift(periods=1)
history
```

```python
(1 - history['retained']/history['previous_months_cust']) * 100
```

```python
avg_purchase = history[history['order_month']<'2018-01']['revenue'].sum()/history[history['order_month']<'2018-01']['orders_volume'].sum()
avg_purchase_freq = history[history['order_month']<'2018-01']['revenue'].sum()/len(set(final[final['order_month']<'2018-01']['customer_unique_id']))
cust_value = avg_purchase * avg_purchase_freq
avg_cust_lifespan = 1/((1 - history['retained']/history['previous_months_cust']).mean() * 100)
CLTV = cust_value * 12 * avg_cust_lifespan
```
